## Rainfall

The file [rain.txt](https://raw.githubusercontent.com/PX2134/extra/master/rain.txt) contains the monthly average rainfalls in millimeters for Cardiff, Berlin and Los Angeles (in that order). Loading the data into a single array, and using this array, plot:

- The monthly rainfall for each city
- The yearly average rainfall for each city
- The average rainfall between all cities for each month
- The seasonal rainfall for each city

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
!wget -q https://raw.githubusercontent.com/PX2134/extra/master/rain.txt
```

```python
rain= np.loadtxt('rain.txt')
```

```python
rain.shape
# 3 rows with 12 columns
```

```python
rain[0,:] # first row with all of the columns (cardiff )
rain[1,:] # berlin
rain[2,:] #LA
```

```python
rain[:,0] # gives rain in january , all cities so all rows but first column
```

```python
for i in range(3):
  print(rain[i,:])

#makes loop that reads all of the data for 3 cities
```

```python
#monthly aveerage rainfall for cities
cities=['Cardiff','Berlin','LA']
months=['jan','feb','mar','apr','may','jun','jul','aug','sep','oct','nov','dec']
for i in range(3):
  plt.plot(months,rain[i,:],marker='o',label=cities[i]) #adding months gives the x points as the months listed

plt.ylabel('rainfall (mm)')
plt.xlabel('months')
plt.legend()
```

```python
#The yearly average rainfall for each city
np.average(rain[0,:]) #average yearly rain fall for cardiff

for i in range(3):
  plt.plot(cities[i],np.average(rain[i,:]),'o')

for i in range(3):
  plt.bar(cities[i],np.average(rain[i,:])) # plots it as a bar chart

plt.ylabel('rainfall (mm)')
plt.xlabel('months')
plt.legend()
```

```python
#The average rainfall between all cities for each month
rain[:,0] # gives january

average= np.zeros(12)
for i in range(12):
  average[i]= np.average(rain[:,i])

plt.plot(months, average,marker='o')
plt.ylabel('rainfall (mm)')
```

```python
np.average(rain,axis=1) # averages over columns

np.average(rain,axis=0) # average over rows
```
